## 3. Slicing: Taking Part of a String

### Start and Stop
Use `text[start:stop]` to take a range of characters.
The `stop` position is not included, which is one of the most important slicing rules to remember.

```python
# Slicing syntax
# Start at 0 an go until but not including 3
# some languages would do with name.subslice(0,3)
name[0:3]  # you read it as start at index 0 and go until but NOT including index 3 (which would have been the 4th character)
```

```python
# begginning 0 is not necessary
name[:3] # so slicing works for sequences
```

```python
name[:4]
```

```python
name[1:3] # so 2nd and 3rd(with index 2) but not the 4th(with index 3)
```

```python
name[1:4] # so 2nd(with index 1), 3rd, 4th, but not the 5th(with index 4)
```

F-strings are a readable way to build strings that include variables and expressions.

### Important Connection
They still create a new string. The main idea is the same: combine existing values into a new piece of text.

```python
# so by putting letter f in front of our string literal
# we get a template of sorts
# then whatever we put into curly braces will get evaluate
newstr = f"2+2={2+2} nice isn't {name[::-1]*3}" # generally we use f-string with variables but we can eval pretty much anything inside curly braces
newstr
```

### Formatting Numbers Inside Curly Braces

Inside an f-string, the part after `:` tells Python how to format a value. This is useful when you want a fixed number of decimal places or when you want numbers to line up neatly.

A pattern such as `.2f` means "show this number as a float with 2 digits after the decimal point". A pattern such as `10.2f` adds a total width of 10 characters as well.

```python
price = 12.34567
# so without affecting price itself we can show formatted output

print(f"Default: {price}")
print(f"Two decimals: {price:.2f}")
print(f"Two decimals: {price:.4f}")
print(f"Width 10, two decimals: {price:10.2f}")
print(f"Zero-padded: {price:08.2f}")
```

```python
seconds=345.24556
width=32 # how much space our formatted value should take up
precision=4 # how many digits to use
print(f'Elapsed {seconds} seconds since start of new era')
print(f'Elapsed {seconds:{width}.{precision}} seconds since start of new era')
# so in the above example we had curly braces inside curly bracers!
print(f'Elapsed {seconds:{9}.{6}} seconds since start of new era')
print(f'Number {75556.2222:{12}.{8}}')
```

## 11. Summary and Common Pitfalls

### Key Ideas to Keep
- a string is a Unicode text sequence
- indexing reads one character and slicing reads a range
- slice `stop` is not included
- strings are immutable, so operations usually return a new string

### Common Beginner Mistakes
- forgetting that Python starts counting at `0`
- mixing strings and numbers without converting first
- expecting a method such as `upper()` to change the original value automatically

These ideas will return in later lectures when we work with lists, files, and larger text-processing tasks.
