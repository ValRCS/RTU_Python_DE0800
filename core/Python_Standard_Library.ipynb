## Datetime module

You do not want to work with datetime by yourself without a library!

````python
# you do not want to calculate dates yourself,
# easy to make mistakes with timezones, Georgian calendar, leap years etc
from datetime import datetime # I import only datetime class from datetime module
# https://docs.python.org/3/library/datetime.html
````

````python
now = datetime.now() #notice datetime needs another datetime
# if we had imported datetime directly
# then we would have had to write datetime.datetime.now() which is no fun )
now
````

````python
time_passed = datetime.now() - now # we have fixed the time passed here in a variable
time_passed
````

````python
print(f"Time passed in seconds {time_passed.seconds}")
# type of time_passed
print(f"Type of time_passed {type(time_passed)}")
````

````python
# how many days have passed?
print(f"Time passed in days {time_passed.days}")
````

````python
# how many total seconds have passed?
print(f"Time passed in seconds {time_passed.total_seconds()}")
````

````python
# we can get individual years, months, days, hours, minutes, seconds
# also date
print(f"Year {now.year}")
print(f"Month {now.month}")
print(f"Day {now.day}")
print(f"Hour {now.hour}")
print(f"Minute {now.minute}")
print(f"Second {now.second}")
print(f"Microsecond {now.microsecond}")
print(f"Date {now.date()}")
print(f"Time {now.time()}")
# day
print(f"Weekday {now.weekday()}") # 0 is Monday, 6 is Sunday
````

## Itertools module - for efficient looping

Itertools module provides a set of functions for working with iterators. It provides many functions that can be used to create and work with iterators.

````python
# let's start with product
# we want to have a Cartesian product of two lists
# we have a list of people
people = ["Alice", "Bob", "Charlie", "David"]
# and we have a list of drinks
drinks = ["Coke", "Beer", "Water"]
prices = [0.99, 4.99]
# we could use two nested loops to get all combinations

# instead it is better to use product
from itertools import product
# so we create a Cartesian product of all the possibilities
all_combinations = list(product(people, drinks, prices))
print("All combinations")
all_combinations
````

````python
## Now permutations
# we have a list of people
people = ["Alice", "Bob", "Charlie", "David"]
# we want to get all permutations of two people
from itertools import permutations
all_permutations = list(permutations(people, 2)) # we want to get all pairs
print("All permutations")
all_permutations
````

````python
## how about combinations, meaning we do not care about order

# we have a list of people
pizza_toppings = ["Pepperoni", "Mushrooms", "Olives", "Onions"]
# we want to get all combinations of two toppings
from itertools import combinations
all_combinations = list(combinations(pizza_toppings, 2)) # we want to get all pairs
print("All combinations of size 2")
all_combinations
````

````python
# now we can also have combinations with replacement
# we will stick pizza toppings
from itertools import combinations_with_replacement
all_combinations = list(combinations_with_replacement(pizza_toppings, 2)) # we want to get all pairs
print("All combinations with replacement of size 2")
all_combinations
````

## Combining zip with cycle

We might have one longer list and one very short list. We can use zip with cycle to iterate over the shorter list multiple times.

````python
# so let's use built in zip - NOT related to zipfile !!!
# and use cycle to repeat the shorter list
from itertools import cycle
# so cycle will loop through whatever you created forever
# we have a list of people
people = ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Hank"]
# and we have a list of drinks
drinks = ["Coke", "Beer", "Water"]
# we want everyone to have a drink so we can cycle them
# we want drinks to be repeated so we use cycle
people_drinks = list(zip(people, cycle(drinks)))
print("People and their drinks")
people_drinks
````

````python
# if you did not know about zip and cycle how would you do it?
# you would use modulo operator
people_drinks = []
for i, person in enumerate(people):
    people_drinks.append((person, drinks[i % len(drinks)]))
# this is much messier and harder to read
print("People and their drinks")
people_drinks
````
